```python
%%R
daily <- read.table("daily.csv", dec='.', sep=',', header=TRUE)
adjprice <- rep(0, length(daily[,1]))

for (i in 1:16143) {
    adjprice[16144 -i] <- daily$Adj.Close[i]
}
prepr <- adjprice[12697:14493]
postpr <- adjprice[14494:15431]
```

```python
%%R
library(stats)
arimapre <- arima(ts(prepr),c(1,0,1))
print(arimapre)
arimapost <- arima(ts(postpr),c(1,1,1))
print(arimapost)
```

```python
%%R
library("forecast")
plot(forecast(object=arimapre, h=1000))
#print(forecast(object=arimapre, h=1000))
#plot(new=T)
#plot(x=1798:2735, y=postpr, axes=F, col=3)
#plot(new=F)
plot(forecast(object=arimapost, h=1000))
print(length(prepr))
print(length(postpr))
print(length(prepr)+length(postpr))
```

```python
%%R
## install.packages("tseries")
library("tseries")
dat1 <- diff(log(prepr))
dat1 <- ts(dat1)
dat1.garch <- garch(dat1)
summary(dat1.garch)    
plot(dat1.garch) 
```

```python
%%R
dat2 <- diff(log(postpr))
dat2 <- ts(dat2)
dat2.garch <- garch(dat2)
summary(dat2.garch)    
plot(dat2.garch) 
```
